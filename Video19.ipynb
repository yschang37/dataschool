# How do I select multiple rows and columns from a pandas DataFrame?

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head(3)
```

```python
ufo.loc[0,:]
```

```python
ufo.loc[[1,3,4],:]
```

```python
ufo.loc[0:2,:]
```

```python
ufo.loc[:,'City']
```

```python
ufo.loc[:,['City','State']]
```

```python
ufo.loc[:,'City':'State']
```

```python
ufo.head(3).drop('Time', axis=1)
```

```python
ufo[ufo.City=="Oakland"]
```

```python
ufo.loc[ufo.City=="Oakland",'State'] #safer
```

```python
ufo.iloc[:,[0,3]]
```

```python
list(range(1,4))
```

```python
ufo.iloc[0:3,:] #iloc for interger position and loc for label
```

```python
ufo[['City','State']]
```

```python
ufo['City','State']
```

```python
ufo[0:2]
```
